```python
import turtle
import time
def main():
    # The first step is to create the screen
    screen = turtle.getscreen()
    # creating background color 
    screen.bgcolor("white")
    # creating title for our flag 
    screen.title("America Flag Final Project")
    america = turtle.Turtle()
    # setting the cursor/turtle speed. high val --> faster turtle
    america.speed(100)
    america.penup()
    # creating shape for the curser/turt
    america.shape("turtle")

    # setting flag dimensions based off standard ratio 
    flag_height = 250
    flag_width = 475

    # starting points for flag
    #starting from the first quadrant - half of flag width & height
    start_x = -237
    start_y = 125

    # For red and white stripes, 13 total for the flag  
    #each strip width will be height/13 = 19.2 approx
    stripe_height = flag_height/13
    stripe_width = flag_width

    # length of one arm of star
    star_size = 10
    #function for the red and white flag stripes

    def draw_rectangle(x, y, height, width, color):
        america.goto(x,y)
        america.pendown()
        america.color(color)
        america.begin_fill()
        america.forward(width)
        america.right(90)
        america.forward(height)
        america.right(90)
        america.forward(width)
        america.right(90)
        america.forward(height)
        america.right(90)
        america.end_fill()
        america.penup()

    def draw_star(x,y,color,length) :
        america.goto(x,y)
        america.setheading(0)
        america.pendown()
        america.begin_fill()
        america.color(color)
        for turn in range(0,5) :
            america.forward(length)
            america.right(144)
            america.forward(length)
            america.right(144)
        america.end_fill()
        america.penup()
    def get_color(color):
        if color== ("red"):
            return ("#FF0000")
        if color == ("blue"):
            return ("#0000FF")
        if color == ("white"):
            return ("#FFFFFF")


    # this function is used to create 13 red and white stripes of flag
    def draw_stripes():
        x = start_x
        y = start_y
        #need 7 red and 6 white stripes - 13 total
        #so first create 6 red and 6 white stripes alternating
        for stripe in range(0,6):
            for color in ["red", "white"]:
                draw_rectangle(x, y, stripe_height, stripe_width, color)
                # decrease value of y by stripe_height
                y = y - stripe_height
        get_color("red")
        get_color("white")

        # create last red stripe
        draw_rectangle(x, y, stripe_height, stripe_width, 'red')
        y = y - stripe_height


    # Creates the blue colored square for the flag
    # height = 7/13 of flag_height
    # width = 0.76 * flag_height
    def draw_flag():
        square_height = (7/13) * flag_height
        square_width = (0.76) * flag_height
        draw_rectangle(start_x, start_y, square_height, square_width, 'blue')
        

     #function to draw 6x5 star rows - 30 of 50 stars
    def draw_six_stars():
        gap_between_stars = 30
        gap_between_lines = stripe_height + 6
        y = 112
        # create 5 rows of stars
        for row in range(0,5) :
            x = -222
            # create 6 stars in each row
            for star in range (0,6) :
                draw_star(x, y, 'white', star_size)
                x = x + gap_between_stars
            y = y - gap_between_lines
        get_color("white")
        
    #function to draw 5x4 star rows - 20 of 50 stars 
    def draw_five_stars():
        gap_between_stars = 30
        gap_between_lines = stripe_height + 6
        y = 100
        # create 4 rows of stars
        for row in range(0,4) :
            x = -206
            # create 5 stars in each row
            for star in range (0,5) :
                draw_star(x, y, 'white', star_size)
                x = x + gap_between_stars
            y = y - gap_between_lines
    draw_stripes()
    draw_flag()
    draw_six_stars()
    draw_five_stars()
    america.hideturtle()
    screen.mainloop()
if __name__== "__main__":
    main()
```
